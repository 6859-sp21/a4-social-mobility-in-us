```python
import pandas as pd
import numpy as np
```

```python
mrc1_df = pd.read_csv('mrc_table1.csv')

# super_opeid
# -99 = late goers (attended college between 23-28)
# -9 = Never attended college up to and including 2013
# -1 = attending a college with insufficient data
# >0 = specific colleges

# We are only interested in specific colleges (filter others)
mrc1_df = mrc1_df[mrc1_df.super_opeid > 0]
mrc1_df
```

```python
mrc10_df = pd.read_csv('mrc_table10.csv')
mrc10_df
```

```python
mrc11_df = pd.read_csv('mrc_table11.csv')
mrc11_df
```

```python
locations_df = pd.read_csv('locations_updated.csv').drop(columns = ['Institution name alias (HD2019)', 'Unnamed: 9'])
locations_df.columns = ['name', 'latitude', 'name2', 'longitude', 'opeid', 'lat12', 'long12', 'opeid12']
locations_df['institution_name'] = locations_df.name.apply(lambda x: x.upper())


locations_df
```

```python
test_same_latlong = locations_df[['latitude', 'longitude']] == locations_df[['lat12', 'long12']].rename(columns = {'lat12':'latitude', 'long12':'longitude'})


test_same_latlong_bool = locations_df.apply(lambda row: abs(row['latitude'] - row['lat12']) < .5 and abs(row['longitude'] - row['long12']) < .5, axis = 1)


# test_same_latlong_bool = test_same_latlong.apply(lambda row: row['latitude'] and row['longitude'], axis = 1)
locations_df = locations_df[test_same_latlong_bool].drop(columns = ['lat12', 'long12', 'name2'])


test_same_opeid_bool = locations_df.apply(lambda row: row['opeid'] == row['opeid12'], axis = 1)
test_same_opeid_bool

locations_df = locations_df[test_same_opeid_bool].drop(columns = ['opeid12'])

locations_df
```

```python
full_df = mrc11_df.merge(locations_df, how = 'left', 
                         left_on = 'institution_name', 
                         right_on = 'institution_name')
print("Not NA count:", len(full_df[-full_df.longitude.isna()]))


full_df = mrc1_df.merge(full_df.drop(columns = ['name']), 
                        how = 'left', 
                        left_on = 'super_opeid', 
                        right_on = 'super_opeid')
print("Not NA count:", len(full_df[-full_df.longitude.isna()]))

full_df = full_df.merge(mrc10_df.drop(columns = ['name', 'czname', 'state', 'multi']), 
                        how = 'left', 
                        left_on = 'super_opeid', 
                        right_on = 'super_opeid')
                        
print("Not NA count:", len(full_df[-full_df.longitude.isna()]))

print('Columns:', full_df[-full_df.longitude.isna()].columns)

full_df = full_df[-full_df.longitude.isna()]

# full_df.to_csv('final_dataset.csv', index = False)

full_df
```
